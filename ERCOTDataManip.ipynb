```python
import glob
import pandas as pd
import os
import csv
import datetime 
os.chdir("/Projects/TPG/Support/")
```

```python
path = os.getcwd()
print(path)
files = os.listdir(path)
files_xlsx = [i for i in files if i[-4:] == 'xlsx']
files_xlsx
```

```python
df = pd.DataFrame()
for f in files_xlsx:
    data = pd.read_excel(f)
    data.columns=['Date', 'COAST', 'EAST', 'FAR_WEST', 'NORTH', 'NORTH_C', 'SOUTHERN',
       'SOUTH_C', 'WEST', 'Total Load']
    df = df.append(data)
    
df.to_csv("2016-2020data.csv")
```

```python
df_totals = df[['Date','Total Load']].reset_index(drop=True)
df_totals['Date'] = df_totals['Date'].replace(to_replace=' DST$', value= '', regex=True)
df_totals['cor'] = pd.to_datetime(df_totals.Date, format='%Y-%m-%d %H:%M:%S.%f', errors='coerce')
df1 = df_totals[~df_totals['Date'].str.contains('24:', na=False)]
```

```python
dfm = df_totals.copy()
dfm = dfm[dfm['Date'].str.contains('24:00', na=False)]
display(dfm)
```

```python
def my_to_datetime(date_str):
    if date_str[11:13] != '24':
        return pd.to_datetime(date_str, format='%m/%d/%Y %H:%M')

    date_str = date_str[0:11] + '00' + date_str[13:]
    return pd.to_datetime(date_str, format='%m/%d/%Y %H:%M') + \
           datetime.timedelta(days=1)
```

```python
##########
dfm['Date'] = dfm.Date.apply(my_to_datetime)
##########
```
